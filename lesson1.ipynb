<h2 style="color:#00796B">üìä Step 4:  Exploratory Data Analysis(EDA) </h2>

```python
custom_palette = {
    'F/yes': '#e91e63',   # female & in a romantic relationship
    'F/no':  '#f8bbd0',   # female & not in a romantic relationship
    'M/yes': '#2196f3',   # male & in a romantic relationship
    'M/no':  '#90caf9'    # male & not in a romantic relationship
}

Data['Gender_and_Romantic Relationship'] = Data['sex'] + "/" + Data['romantic']

sns.countplot(data=Data,x='studytime' , hue = "Gender_and_Romantic Relationship" , palette = custom_palette )
plt.ylabel('Number of Students')
plt.xlabel('Study Hours')
plt.title('Study Time by Gender and Romantic Relationship')
plt.show()
```

```python
# List of numerical features to visualize
num_features = [
    'G1', 'G2', 'G3',         # Grades
    'age', 'studytime', 'traveltime', 'failures', 'absences',
    'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health'
]

# Set plot style
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(16, 18))

# Loop through features and plot
for i, feature in enumerate(num_features):
    plt.subplot(5, 3, i + 1)
    Data[feature].hist(bins=15, edgecolor='black')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## üìä Understanding the Histograms in EDA

Histograms help us visualize the **distribution** of numerical data and uncover useful patterns, such as skewness, modality, or the presence of outliers. Below is a breakdown of what each histogram reveals about students‚Äô behavior, lifestyle, and academic performance:

---

### üéì Academic Performance

- **G1, G2, G3**:  
  These histograms show how students are distributed across different grade levels. A **left-skewed** distribution might indicate grade inflation, while a **right-skewed** pattern may reveal underperformance.

---

### üß† Effort & Commitment

- **studytime**:  
  Helps us see how many students study less than 2 hours vs. more than 10 hours per week. Can highlight motivation levels or time constraints.

- **traveltime**:  
  Shows how far students live from school. Peaks at lower times may mean most students live nearby.

- **failures**:  
  Reveals how common academic struggles are. A spike at 0 indicates many students haven't failed, but outliers (3 or more) may need additional support.

- **absences**:  
  Tracks school attendance. Can uncover chronic absenteeism or discipline issues affecting learning.

---

### üë®‚Äçüë©‚Äçüëß‚Äçüë¶ Family & Social Factors

- **famrel**:  
  Distribution of family relationship quality. A central peak around 4‚Äì5 implies mostly good relations, which might correlate positively with performance.

- **freetime**, **goout**:  
  Show how students balance social life and school. Useful for exploring whether excessive leisure affects grades.

---

### üç∑ Lifestyle Habits

- **Dalc**, **Walc**:  
  Important for understanding alcohol consumption patterns. Spikes at lower levels are ideal, but higher peaks during weekends (`Walc`) may indicate behavioral trends.

---

### ‚ù§Ô∏è Health & Age

- **health**:  
  Self-assessed health levels; usually centered unless specific issues dominate.

- **age**:  
  Should follow a bell-shaped distribution based on school-age norms (mostly 15‚Äì19). Outliers could point to non-traditional learners or repeating students.

---

## üîç Insights & Use Cases

- Identify **at-risk students** using combinations of high absences, low grades, and low studytime.
- Design **interventions** targeted at lifestyle or academic behavior extremes.
- Tailor **curriculum plans** around common student profiles revealed by the histograms.

---

‚úÖ Visuals help uncover patterns:
- Do students in relationships study more or less?
- Are there gender differences in study behavior?


 <h2 style="color:#5D4037">üîó Step 5: Correlation Matrix</h2>

```python
#   c) Correlation Matrix (numeric-only)
corr_matrix = Data.corr(numeric_only=True)
plt.figure(figsize=(20, 8))
plt.gcf().set_facecolor('#FFFDD0') 
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix without Categorical Features')
plt.show()
```

‚úÖ This reveals:
- Which features are strongly related to final grade (`G3`)  
- Useful inputs for later modeling (e.g. `G1`, `G2`)
